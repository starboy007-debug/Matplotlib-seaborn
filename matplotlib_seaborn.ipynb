# bar plot

```python
x = [2000,2001,2002,2003,2004,2005]
y = [12345,15432,17811,10923,19000,14001]
xpos = np.arange(len(x))
plt.bar(xpos,y,color='r',edgecolor='k',label='klkl',alpha=0.5)
plt.xticks(xpos,x,rotation=30)
plt.title('bargraph')
plt.xlabel('year')
plt.ylabel('sales')
plt.legend()
plt.show()
```

```python
x = [2000,2001,2002,2003,2004,2005]
y1 = [12345,15432,17811,10923,19000,14001]
y2 = [22345,25432,27811,20923,29000,24001]
y3 = [32345,35432,37811,30923,39000,34001]
xpos = np.arange(len(x))
width=0.25
plt.bar(xpos-width,y1,color='r',edgecolor='k',label='kl',alpha=0.5)
plt.bar(xpos,y2,color='b',edgecolor='k',label='klk',alpha=0.5)
plt.bar(xpos+width,y3,color='g',edgecolor='k',label='klkl',alpha=0.5)
plt.xticks(xpos,x,rotation=30)
plt.title('bargraph')
plt.xlabel('year')
plt.ylabel('sales')
plt.legend()
plt.show()
```

# Animation

```python
%matplotlib notebook
from matplotlib import animation
```

```python
fig = plt.figure()
ax1 = plt.subplot(111)
line, = ax1.plot([],[])

ax1.set_xlim(0,2)
ax1.set_ylim(-1.5,1.5)

def init():
    ax1.set_xlabel('x')
    ax1.set_ylabel('sin(x)')
    ax1.set_title('sin curve')
    line.set_data([],[])
    return line,

def animate(i):
    x = np.linspace(0,2,1000)
    y = np.sin(2*np.pi*(x-0.01*i))
    line.set_data(x,y)
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=100, blit=True, repeat=True)

plt.show()
```

```python
anim.save('basic_animation.mp4', fps=30)
```
